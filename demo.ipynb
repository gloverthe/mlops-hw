```python
import requests
import monitor
import os
from pathlib import Path
import sys

# Test the API 

url = "http://localhost:8000/predict"
payload = {
    "instances": [
        [5.1, 3.5, 1.4, 0.2],
        [6.2, 3.4, 5.4, 2.3]
    ]
}

resp = requests.post(url, json=payload, timeout=10)
resp.raise_for_status()
print(resp.json())


log_file = Path('logs') / 'api.log'

# View API monitoring stats
if log_file.exists():
    api_monitor = monitor.monitor(log_file=log_file)
    api_monitor.plot_stats()
    api_monitor.display_summary()
```
